```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('training.csv')
```

```python
data.describe()
```

```python
data.columns
```

```python
drop_useless_columns = ['comp1_rate','comp1_inv',
       'comp1_rate_percent_diff', 'comp2_rate', 'comp2_inv',
       'comp2_rate_percent_diff', 'comp3_rate', 'comp3_inv',
       'comp3_rate_percent_diff', 'comp4_rate', 'comp4_inv',
       'comp4_rate_percent_diff', 'comp5_rate', 'comp5_inv',
       'comp5_rate_percent_diff', 'comp6_rate', 'comp6_inv',
       'comp6_rate_percent_diff', 'comp7_rate', 'comp7_inv',
       'comp7_rate_percent_diff', 'comp8_rate', 'comp8_inv',
       'comp8_rate_percent_diff']
data.drop(drop_useless_columns,axis=1,inplace=True)
```

```python
data.describe()
```

```python
null_value_columns = data.columns[data.isnull().any()]
data[null_value_columns].isnull().sum()
```

```python
##Dropping further high null value columns
new_drop_columns = ['visitor_hist_starrating','visitor_hist_adr_usd','gross_bookings_usd',
                    'srch_query_affinity_score','prop_location_score2','orig_destination_distance']
data.drop(new_drop_columns,axis=1,inplace=True)
```

```python
data['prop_review_score'].fillna(data['prop_review_score'].mean(),inplace=True)
```

```python
##Building a new feature
conditions = [
    (data['click_bool'] == 0) & (data['booking_bool'] == 0),
    (data['click_bool'] == 0) & (data['booking_bool'] == 1),
    (data['click_bool'] == 1) & (data['booking_bool'] == 0),
    (data['click_bool'] == 1) & (data['booking_bool'] == 1)
    ]
values = [0,-1,-1,1]
data['Output']=np.select(conditions,values,default=0)
```

```python
data.info()
```

```python
#correlation matrix 
corr = data.corr()
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['xtick.color']='white'
mpl.rcParams['ytick.color']='white'
plt.figure(figsize=(20,20))
sns.heatmap(corr,xticklabels=data.columns,yticklabels=data.columns,cmap = 'RdYlGn')
```

```python
#Dropping the date field
data.drop(['date_time'],axis=1,inplace=True)
```

```python
data.drop(['click_bool','booking_bool','Output'],axis=1,inplace=True)
```

```python
##As no correlation, we are going to do PCA
## first use standard scaler then PCA
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
data_pca = data[0:100000]
scaled_data = StandardScaler()
```

```python
x =scaled_data.fit_transform(data_pca)
```

```python
#initiating PCA
pca = PCA()
pca.fit(x)
pca_data = pca.transform(x)
```

```python
per_variation = pca.explained_variance_
labels = [str(x) for x in range(1,len(per_variation)+1)]
plt.bar(x=range(1,len(per_variation)+1),height=per_variation,tick_label = labels)
plt.ylabel('percentage')
plt.xlabel('components')
plt.show
```
